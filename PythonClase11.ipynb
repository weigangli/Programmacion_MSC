```python
import requests
from bs4 import BeautifulSoup

url="https://informatica.ucm.es/informatica/profesores-y-tutorias"
req=requests.get(url)
soup=BeautifulSoup(req.text,"html.parser")
soup.title
```

```python
for link in soup.find_all('a'):
    print(link.text)
```

```python
print(soup.body.prettify())
```

```python
url=soup.find('iframe').get('src')
url
```

```python
req=requests.get(url)
soup=BeautifulSoup(req.text,"html.parser")
soup.body.prettify()
```

```python
headers=soup.find_all('th')
for header in headers:
    print(header.text)
```

```python
rows=soup.find_all('tr')

data=[]

for row in rows:
    cells=row.find_all('td')
    if len(cells)>2:
        data.append((cells[0].text,
                     cells[1].text,
                     cells[2].text,
                     cells[3].text,
                     cells[4].text))
data
```

```python
import pandas as pd

df=pd.DataFrame(data,columns=(headers[0].text.title(),
                              headers[1].text,
                              headers[2].text,
                              headers[3].text,
                              headers[4].text[:3]))
df
```

```python
dfd=df.sort_values(['Dpto.','Profesor'])
dfd
```

```python
dfd.to_csv('profesores.csv')
```
